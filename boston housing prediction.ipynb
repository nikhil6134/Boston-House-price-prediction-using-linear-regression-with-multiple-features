```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#normalise the data 

u = np.mean(X,axis = 0)
std = np.std(X,axis = 0)

X = (X-u)/std

d = pd.DataFrame(X)

d
```

```python
# X  - Matrix (m x n m exmples and n features)
# x- vector (Single example with n features)

ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))


def hypothesis(x,theta):
    y_ = 0.0
    n = x.shape[0]
    
    for i in range(n):
        y_ += (theta[i]*x[i])
        
    return y_

def error(X,Y,theta):
    e = 0.0
    m = X.shape[0]
    
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        
        e+= (Y[i]- y_)**2
        
        
    return e/m

def gradient(X,Y,theta):
    m,n = X.shape
    
    
    grad = np.zeros((n,))
    
    # for all values of j
    for j in range(n):
        
        #sum over all examples
        
        for i in range(m):
            y_ = hypothesis(X[i],theta)
            
            grad[j] += (y_-Y[i])*X[i][j]
            
    return grad/m


def gradient_descent(X,Y,learning_rate = 0.1,max_echops = 300):
    m,n = X.shape
    theta = np.zeros((n,))
    error_list = []
    
    for i in range(max_echops):
        e = error(X,Y,theta)
        error_list.append(e)
        
        #gradient descent
        grad = gradient(X,Y,theta)
        for j in range(n):
            theta[j] = theta[j] - learning_rate*grad[j]
            
            
    return theta,error_list


            
        
        
    
        
        
```

```python
import time

start = time.time()

theta,error_list = gradient_descent(X,Y)
end = time.time()

print("Time taken is",end-start)
```

```python
def r2_score(Y,y_):
    num = np.sum((Y-y_)**2)
    denum = np.sum((Y-Y.mean())**2)
    score= (1-num/denum)
    return score*100

```

```python
# lets vectorize the code so it will impelement fast

def hypothesis(X,theta):
    
    return np.dot(X,theta)

def error(X,Y, theta):
    e = 0.0
    m= X.shape[0]
    
    y_ = hypothesis(X,theta)
    e = np.sum((Y-y_)**2)
    
    
    return e/m

def gradient(X,Y,theta):
    
    n = X.shape[1]
    m= X.shape[0]
    

    y_ = hypothesis(X,theta)
    
    grad = np.dot(X.T,(y_-Y))
    
    return grad/m


    
def gradient_descent(X,Y,learning_rate=0.1,max_ephocs = 300):
    
    n = X.shape[1]
    
    theta = np.zeros((n,))
    error_list = []
    for i in range(max_ephocs):
        
        e = error(X,Y,theta)
        error_list.append(e)
        
        grad = gradient(X,Y,theta)
        
        
            
        theta = theta - learning_rate*grad
            
            
    return theta,error_list




        
    
```

```python
import time
start = time.time()
theta,error_list = gradient_descent(X,Y)
end = time.time()

print("the time taken is:",end-start)
```
